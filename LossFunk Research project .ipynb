```python
# Enhanced RL Reward Comparison Framework
# Complete implementation with GSM8K focus, Sophisticated Rewards, Hybrid Schedules, Training Dynamics

import re
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple, Any, Optional
from dataclasses import dataclass
from collections import defaultdict
import json
import pandas as pd
from datasets import load_dataset, Dataset
from unsloth import FastLanguageModel, is_bfloat16_supported
from trl import GRPOConfig, GRPOTrainer
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch.nn.functional as F
from difflib import SequenceMatcher
import random

# Configuration
MAX_SEQ_LENGTH = 1024
LORA_RANK = 8
EVAL_BATCH_SIZE = 16

@dataclass
class TrainingMetrics:
    """Store training metrics for comparison."""
    step: int
    accuracy: float
    avg_reward: float
    reward_variance: float
    perplexity: float
    loss: float
    grad_norm: float
    reward_components: Dict[str, float] = None

@dataclass
class ExperimentResults:
    """Store complete experiment results."""
    name: str
    metrics: List[TrainingMetrics]
    final_accuracy: float
    convergence_step: int
    training_stability: float
    reward_distribution: List[float]
    domain: str = "gsm8k"
    reward_type: str = "hard"

# System prompt for GSM8K
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
Show your step-by-step mathematical reasoning here.
</reasoning>
<answer>
Provide the final numerical answer here.
</answer>
"""

def setup_model():
    """Initialize and configure the model with LoRA."""
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name="Qwen/Qwen2.5-3B-Instruct",
        max_seq_length=MAX_SEQ_LENGTH,
        load_in_4bit=True,
        fast_inference=True,
        max_lora_rank=LORA_RANK,
        gpu_memory_utilization=0.5,
    )
    
    model = FastLanguageModel.get_peft_model(
        model,
        r=LORA_RANK,
        target_modules=[
            "q_proj", "k_proj", "v_proj", "o_proj",
            "gate_proj", "up_proj", "down_proj",
        ],
        lora_alpha=LORA_RANK,
        use_gradient_checkpointing="unsloth",
        random_state=3407,
    )
    
    return model, tokenizer

def extract_xml_answer(text: str) -> str:
    """Extract answer from XML format."""
    if "<answer>" not in text or "</answer>" not in text:
        return ""
    answer = text.split("<answer>")[-1]
    answer = answer.split("</answer>")[0]
    return answer.strip()

def extract_hash_answer(text: str) -> str | None:
    """Extract answer from hash format (####)."""
    if "####" not in text:
        return None
    return text.split("####")[1].strip()

def get_gsm8k_questions(split="train", limit=None) -> Dataset:
    """Load and preprocess GSM8K dataset with actual datapoints."""
    data = load_dataset('openai/gsm8k', 'main')[split]
    if limit:
        data = data.select(range(min(limit, len(data))))
    
    def process_example(example):
        # Extract the numerical answer from the solution
        answer = extract_hash_answer(example['answer'])
        
        return {
            'prompt': [
                {'role': 'system', 'content': SYSTEM_PROMPT},
                {'role': 'user', 'content': example['question']}
            ],
            'answer': answer,
            'question': example['question'],
            'full_solution': example['answer'],
            'domain': 'gsm8k'
        }
    
    processed_data = data.map(process_example)
    return processed_data

# ==================== ENHANCED REWARD TRACKER ====================

class EnhancedRewardTracker:
    """Enhanced reward tracker with component-wise tracking."""
    def __init__(self):
        self.rewards = []
        self.accuracies = []
        self.perplexities = []
        self.reward_components = defaultdict(list)
        self.step = 0
        self.training_phase = "initial"
        self.phase_transitions = []
        self.gradient_norms = []
        self.learning_rates = []
        self.step_rewards = []  # Track rewards per step for dynamics
    
    def log_rewards(self, rewards: List[float], accuracy: float, perplexity: float = None, 
                   components: Dict[str, List[float]] = None, grad_norm: float = None, lr: float = None):
        self.rewards.extend(rewards)
        self.step_rewards.append(rewards)
        self.accuracies.append(accuracy)
        if perplexity:
            self.perplexities.append(perplexity)
        if grad_norm:
            self.gradient_norms.append(grad_norm)
        if lr:
            self.learning_rates.append(lr)
        
        if components:
            for comp_name, comp_values in components.items():
                self.reward_components[comp_name].extend(comp_values)
        
        self.step += 1
    
    def set_training_phase(self, phase: str):
        if phase != self.training_phase:
            self.phase_transitions.append((self.step, phase))
            self.training_phase = phase

# Global tracker instance
reward_tracker = EnhancedRewardTracker()

# ==================== SOPHISTICATED REWARD FUNCTIONS ====================

def calculate_perplexity(model, tokenizer, text: str) -> float:
    """Calculate perplexity of generated text."""
    try:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        if torch.cuda.is_available():
            inputs = {k: v.cuda() for k, v in inputs.items()}
        
        with torch.no_grad():
            outputs = model(**inputs, labels=inputs["input_ids"])
            loss = outputs.loss
            perplexity = torch.exp(loss).item()
        
        return min(perplexity, 1000.0)
    except:
        return 1000.0

def sophisticated_perplexity_reward_func(prompts, completions, answer, **kwargs) -> list[float]:
    """Sophisticated continuous reward with multiple components."""
    model = kwargs.get('model')
    tokenizer = kwargs.get('tokenizer')
    
    if not model or not tokenizer:
        return [0.5] * len(completions)  # Fallback
    
    responses = [completion[0]['content'] for completion in completions]
    extracted_responses = [extract_xml_answer(r) for r in responses]
    
    rewards = []
    components = {
        'correctness': [],
        'perplexity': [],
        'consistency': [],
        'reasoning_quality': []
    }
    
    for i, (response, extracted, true_answer) in enumerate(zip(responses, extracted_responses, answer)):
        # 1. Enhanced Perplexity Component
        full_perplexity = calculate_perplexity(model, tokenizer, response)
        
        # Separate reasoning and answer perplexities
        reasoning_text = ""
        answer_text = ""
        if "<reasoning>" in response and "</reasoning>" in response:
            reasoning_text = response.split("<reasoning>")[1].split("</reasoning>")[0]
        if "<answer>" in response and "</answer>" in response:
            answer_text = response.split("<answer>")[1].split("</answer>")[0]
        
        reasoning_perplexity = calculate_perplexity(model, tokenizer, reasoning_text) if reasoning_text else full_perplexity
        answer_perplexity = calculate_perplexity(model, tokenizer, answer_text) if answer_text else full_perplexity
        
        # Weighted perplexity score
        perplexity_reward = (
            0.4 * np.exp(-full_perplexity / 100.0) +
            0.3 * np.exp(-reasoning_perplexity / 80.0) +
            0.3 * np.exp(-answer_perplexity / 60.0)
        )
        components['perplexity'].append(perplexity_reward)
        
        # 2. Enhanced Correctness Component
        correctness_reward = 0.0
        if extracted and true_answer:
            if extracted == true_answer:
                correctness_reward = 1.0
            else:
                try:
                    pred_num = float(extracted.replace(',', '').replace('$', ''))
                    true_num = float(true_answer.replace(',', '').replace('$', ''))
                    
                    # Multiple similarity metrics
                    relative_error = abs(pred_num - true_num) / max(abs(true_num), 1.0)
                    magnitude_similarity = 1.0 / (1.0 + abs(np.log10(max(abs(pred_num), 0.01)) - 
                                                               np.log10(max(abs(true_num), 0.01))))
                    
                    # Order of magnitude check
                    order_of_magnitude_correct = abs(np.log10(max(abs(pred_num), 0.01)) - 
                                                   np.log10(max(abs(true_num), 0.01))) < 1
                    
                    correctness_reward = (
                        0.6 * np.exp(-relative_error) +
                        0.25 * magnitude_similarity +
                        0.15 * (1.0 if order_of_magnitude_correct else 0.0)
                    )
                    
                except:
                    # Enhanced string similarity
                    sequence_similarity = SequenceMatcher(None, extracted, true_answer).ratio()
                    word_overlap = len(set(extracted.split()) & set(true_answer.split())) / max(len(set(true_answer.split())), 1)
                    correctness_reward = 0.7 * sequence_similarity + 0.3 * word_overlap
        
        components['correctness'].append(correctness_reward)
        
        # 3. Reasoning Quality Component
        reasoning_quality = 0.0
        if reasoning_text:
            # Length appropriateness
            word_count = len(reasoning_text.split())
            length_score = 1.0 if 20 <= word_count <= 200 else max(0, 1 - abs(word_count - 100) / 100)
            
            # Step-by-step indicators
            step_indicators = len(re.findall(r'step|first|second|third|then|next|finally', reasoning_text.lower()))
            step_score = min(1.0, step_indicators / 3)
            
            # Mathematical reasoning indicators
            math_indicators = len(re.findall(r'\d+|\+|\-|\*|\/|\=|calculate|multiply|divide|add|subtract', reasoning_text))
            math_score = min(1.0, math_indicators / 5)
            
            reasoning_quality = 0.4 * length_score + 0.3 * step_score + 0.3 * math_score
        
        components['reasoning_quality'].append(reasoning_quality)
        
        # 4. Consistency Component
        consistency_reward = 0.0
        if reasoning_text and answer_text:
            # Check if numbers in reasoning match the answer
            reasoning_numbers = re.findall(r'\d+(?:\.\d+)?', reasoning_text)
            answer_numbers = re.findall(r'\d+(?:\.\d+)?', answer_text)
            
            if reasoning_numbers and answer_numbers:
                try:
                    final_reasoning_num = float(reasoning_numbers[-1])
                    final_answer_num = float(answer_numbers[-1])
                    consistency_reward = 1.0 if abs(final_reasoning_num - final_answer_num) < 0.01 else 0.5
                except:
                    consistency_reward = 0.3  # Some consistency points for having numbers
        
        components['consistency'].append(consistency_reward)
        
        # Combine all components with weights
        total_reward = (
            0.4 * correctness_reward +
            0.25 * perplexity_reward +
            0.2 * reasoning_quality +
            0.15 * consistency_reward
        )
        
        rewards.append(total_reward)
    
    # Calculate accuracy for tracking
    correct = [r == a for r, a in zip(extracted_responses, answer)]
    accuracy = sum(correct) / len(correct)
    avg_perplexity = np.mean([calculate_perplexity(model, tokenizer, r) for r in responses])
    
    # Log for tracking
    reward_tracker.log_rewards(rewards, accuracy, avg_perplexity, components)
    
    print(f"Step {reward_tracker.step}: Accuracy = {accuracy:.3f}, Avg Reward = {np.mean(rewards):.3f}, Avg Perplexity = {avg_perplexity:.2f}")
    
    return rewards

# ==================== HARD REWARD FUNCTIONS ====================

def hard_correctness_reward_func(prompts, completions, answer, **kwargs) -> list[float]:
    """Traditional binary reward: 1.0 for correct, 0.0 for incorrect."""
    responses = [completion[0]['content'] for completion in completions]
    extracted_responses = [extract_xml_answer(r) for r in responses]
    
    # Calculate accuracy
    correct = [r == a for r, a in zip(extracted_responses, answer)]
    accuracy = sum(correct) / len(correct)
    
    # Binary rewards
    rewards = [1.0 if is_correct else 0.0 for is_correct in correct]
    
    # Log for tracking
    reward_tracker.log_rewards(rewards, accuracy)
    
    print(f"Step {reward_tracker.step}: Accuracy = {accuracy:.3f}, Avg Reward = {np.mean(rewards):.3f}")
    
    return rewards

def hard_format_reward_func(completions, **kwargs) -> list[float]:
    """Binary reward for correct XML format."""
    responses = [completion[0]["content"] for completion in completions]
    has_format = [("<reasoning>" in r and "</reasoning>" in r and 
                  "<answer>" in r and "</answer>" in r) for r in responses]
    return [0.2 if has_fmt else 0.0 for has_fmt in has_format]

# ==================== HYBRID REWARD SCHEDULES ====================

class HybridRewardScheduler:
    """Implements hybrid reward schedules that change over training."""
    
    def __init__(self, schedule_type="continuous_to_hard", transition_steps=[5, 10]):
        self.schedule_type = schedule_type
        self.transition_steps = transition_steps
        self.current_step = 0
        
    def get_reward_weights(self, step: int) -> Dict[str, float]:
        """Get reward weights based on current training step."""
        self.current_step = step
        
        if self.schedule_type == "continuous_to_hard":
            # Start with continuous rewards, gradually shift to hard rewards
            if step < self.transition_steps[0]:
                return {"continuous": 1.0, "hard": 0.0}  # Pure continuous
            elif step < self.transition_steps[1]:
                # Linear transition
                progress = (step - self.transition_steps[0]) / (self.transition_steps[1] - self.transition_steps[0])
                return {"continuous": 1.0 - progress, "hard": progress}
            else:
                return {"continuous": 0.0, "hard": 1.0}  # Pure hard
                
        elif self.schedule_type == "hard_to_continuous":
            # Start with hard rewards, gradually shift to continuous
            if step < self.transition_steps[0]:
                return {"continuous": 0.0, "hard": 1.0}  # Pure hard
            elif step < self.transition_steps[1]:
                # Linear transition
                progress = (step - self.transition_steps[0]) / (self.transition_steps[1] - self.transition_steps[0])
                return {"continuous": progress, "hard": 1.0 - progress}
            else:
                return {"continuous": 1.0, "hard": 0.0}  # Pure continuous
                
        elif self.schedule_type == "alternating":
            # Alternate between hard and continuous every few steps
            cycle_length = self.transition_steps[0]
            if (step // cycle_length) % 2 == 0:
                return {"continuous": 1.0, "hard": 0.0}
            else:
                return {"continuous": 0.0, "hard": 1.0}
                
        else:  # "balanced"
            return {"continuous": 0.5, "hard": 0.5}

def hybrid_reward_func(prompts, completions, answer, **kwargs) -> list[float]:
    """Hybrid reward function that combines hard and continuous rewards based on schedule."""
    scheduler = kwargs.get('scheduler')
    model = kwargs.get('model')
    tokenizer = kwargs.get('tokenizer')
    current_step = reward_tracker.step
    
    if not scheduler:
        # Default to balanced approach
        weights = {"continuous": 0.5, "hard": 0.5}
    else:
        weights = scheduler.get_reward_weights(current_step)
    
    # Get both types of rewards
    hard_rewards = hard_correctness_reward_func(prompts, completions, answer, **kwargs)
    continuous_rewards = sophisticated_perplexity_reward_func(prompts, completions, answer, **kwargs)
    
    # Combine based on weights
    hybrid_rewards = [
        weights["hard"] * hard + weights["continuous"] * cont
        for hard, cont in zip(hard_rewards, continuous_rewards)
    ]
    
    # Update training phase for tracking
    if weights["hard"] > 0.7:
        reward_tracker.set_training_phase("hard_dominant")
    elif weights["continuous"] > 0.7:
        reward_tracker.set_training_phase("continuous_dominant")
    else:
        reward_tracker.set_training_phase("balanced")
    
    print(f"Step {current_step}: Hybrid weights - Hard: {weights['hard']:.2f}, Continuous: {weights['continuous']:.2f}")
    
    return hybrid_rewards

# ==================== TRAINING CONFIGURATIONS ====================

def get_hard_reward_config():
    """Configuration for hard reward training."""
    return {
        'name': 'Hard_Rewards',
        'reward_funcs': [
            hard_correctness_reward_func,
            hard_format_reward_func,
        ],
        'description': 'Binary rewards: 1.0 for correct answers, 0.0 for incorrect',
        'reward_type': 'hard'
    }

def get_continuous_reward_config():
    """Configuration for continuous reward training."""
    return {
        'name': 'Continuous_Rewards',
        'reward_funcs': [
            sophisticated_perplexity_reward_func,
        ],
        'description': 'Continuous rewards based on perplexity and partial correctness',
        'reward_type': 'continuous'
    }

def get_hybrid_reward_config(schedule_type="continuous_to_hard"):
    """Configuration for hybrid reward training."""
    scheduler = HybridRewardScheduler(schedule_type, transition_steps=[3, 7])
    
    return {
        'name': f'Hybrid_Rewards_{schedule_type}',
        'reward_funcs': [hybrid_reward_func],
        'description': f'Hybrid rewards with {schedule_type} schedule',
        'reward_type': 'hybrid',
        'scheduler': scheduler
    }

def setup_training_config(experiment_name: str):
    """Configure GRPO training parameters."""
    return GRPOConfig(
        use_vllm=True,
        learning_rate=5e-6,
        adam_beta1=0.9,
        adam_beta2=0.99,
        weight_decay=0.1,
        warmup_ratio=0.1,
        lr_scheduler_type="cosine",
        optim="adamw_8bit",
        logging_steps=1,
        bf16=is_bfloat16_supported(),
        fp16=not is_bfloat16_supported(),
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        num_generations=2,
        max_prompt_length=256,
        max_completion_length=300,
        max_steps=10,  
        save_steps=5,
        eval_steps=2,
        max_grad_norm=0.1,
        report_to="none",
        output_dir=f"outputs_{experiment_name}",
        logging_dir=f"logs_{experiment_name}",
    )

# ==================== EVALUATION FUNCTIONS ====================

def evaluate_model(model, tokenizer, eval_dataset, num_samples=100):
    """Evaluate model accuracy on test set."""
    correct = 0
    total = 0
    perplexities = []
    
    eval_data = eval_dataset.select(range(min(num_samples, len(eval_dataset))))
    
    for example in eval_data:
        try:
            # Generate response
            messages = example['prompt']
            inputs = tokenizer.apply_chat_template(
                messages, 
                return_tensors="pt",
                add_generation_prompt=True
            )
            
            if torch.cuda.is_available():
                inputs = inputs.cuda()
            
            with torch.no_grad():
                outputs = model.generate(
                    inputs,
                    max_new_tokens=200,
                    temperature=0.7,
                    do_sample=True,
                    pad_token_id=tokenizer.eos_token_id
                )
            
            response = tokenizer.decode(outputs[0][len(inputs[0]):], skip_special_tokens=True)
            extracted = extract_xml_answer(response)
            
            # Check correctness
            if extracted == example['answer']:
                correct += 1
            
            # Calculate perplexity
            perplexity = calculate_perplexity(model, tokenizer, response)
            perplexities.append(perplexity)
            
            total += 1
            
        except Exception as e:
            print(f"Evaluation error: {e}")
            continue
    
    accuracy = correct / total if total > 0 else 0.0
    avg_perplexity = np.mean(perplexities) if perplexities else float('inf')
    
    return accuracy, avg_perplexity

# ==================== EXPERIMENT RUNNER ====================

def run_experiment(config: Dict, train_dataset: Dataset, eval_dataset: Dataset) -> ExperimentResults:
    """Run a single experiment with given configuration."""
    print(f"\n{'='*50}")
    print(f"Running experiment: {config['name']}")
    print(f"Description: {config['description']}")
    print(f"{'='*50}")
    
    # Reset tracker
    global reward_tracker
    reward_tracker = EnhancedRewardTracker()
    
    # Setup model
    model, tokenizer = setup_model()
    
    # Setup training
    training_args = setup_training_config(config['name'])
    
    # Add model and tokenizer to kwargs for rewards
    def create_reward_func_with_context(reward_func):
        def wrapper(*args, **kwargs):
            kwargs['model'] = model
            kwargs['tokenizer'] = tokenizer
            if 'scheduler' in config:
                kwargs['scheduler'] = config['scheduler']
            return reward_func(*args, **kwargs)
        return wrapper
    
    reward_funcs = [create_reward_func_with_context(rf) for rf in config['reward_funcs']]
    
    trainer = GRPOTrainer(
        model=model,
        processing_class=tokenizer,
        reward_funcs=reward_funcs,
        args=training_args,
        train_dataset=train_dataset,
    )
    
    # Train model
    trainer.train()
    
    # Final evaluation
    final_accuracy, final_perplexity = evaluate_model(model, tokenizer, eval_dataset)
    
    # Calculate metrics
    metrics = []
    for i in range(len(reward_tracker.accuracies)):
        step = (i + 1) * training_args.eval_steps
        avg_reward = np.mean(reward_tracker.step_rewards[i]) if i < len(reward_tracker.step_rewards) else 0.0
        reward_var = np.var(reward_tracker.step_rewards[i]) if i < len(reward_tracker.step_rewards) else 0.0
        
        # Get component averages if available
        component_avgs = {}
        for comp_name, comp_values in reward_tracker.reward_components.items():
            if len(comp_values) > i * 8:
                component_avgs[comp_name] = np.mean(comp_values[i*8:(i+1)*8])
        
        metrics.append(TrainingMetrics(
            step=step,
            accuracy=reward_tracker.accuracies[i],
            avg_reward=avg_reward,
            reward_variance=reward_var,
            perplexity=reward_tracker.perplexities[i] if i < len(reward_tracker.perplexities) else 0.0,
            loss=0.0,  # Would need to extract from trainer logs
            grad_norm=reward_tracker.gradient_norms[i] if i < len(reward_tracker.gradient_norms) else 0.0,
            reward_components=component_avgs
        ))
    
    # Calculate convergence step
    convergence_step = len(metrics) * training_args.eval_steps
    if len(reward_tracker.accuracies) > 3:
        for i in range(3, len(reward_tracker.accuracies)):
            recent_improvement = reward_tracker.accuracies[i] - reward_tracker.accuracies[i-3]
            if recent_improvement < 0.01:  # Less than 1% improvement over 3 steps
                convergence_step = i * training_args.eval_steps
                break
    
    # Calculate training stability
    training_stability = 1.0 / (1.0 + np.var(reward_tracker.rewards)) if reward_tracker.rewards else 0.0
    
    return ExperimentResults(
        name=config['name'],
        metrics=metrics,
        final_accuracy=final_accuracy,
        convergence_step=convergence_step,
        training_stability=training_stability,
        reward_distribution=reward_tracker.rewards,
        domain="gsm8k",
        reward_type=config.get('reward_type', 'unknown')
    )

# ==================== TRAINING DYNAMICS VISUALIZATION ====================

def visualize_training_dynamics(results: List[ExperimentResults]):
    """Create comprehensive training dynamics visualizations."""
    fig = plt.figure(figsize=(20, 15))
    
    # Create a 3x3 grid of subplots
    gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
    
    # Define colors for different experiment types
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
    
    # 1. Accuracy over time
    ax1 = fig.add_subplot(gs[0, 0])
    for i, result in enumerate(results):
        steps = [m.step for m in result.metrics]
        accuracies = [m.accuracy for m in result.metrics]
        ax1.plot(steps, accuracies, marker='o', label=result.name, 
                linewidth=2, color=colors[i % len(colors)])
        
        # Mark phase transitions for hybrid methods
        if hasattr(reward_tracker, 'phase_transitions') and reward_tracker.phase_transitions:
            for transition_step, phase in reward_tracker.phase_transitions:
                ax1.axvline(x=transition_step, color=colors[i % len(colors)], 
                           linestyle='--', alpha=0.5)
    
    ax1.set_xlabel('Training Steps')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Accuracy Evolution Over Time')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Average reward over time
    ax2 = fig.add_subplot(gs[0, 1])
    for i, result in enumerate(results):
        steps = [m.step for m in result.metrics]
        avg_rewards = [m.avg_reward for m in result.metrics]
        ax2.plot(steps, avg_rewards, marker='s', label=result.name, 
                linewidth=2, color=colors[i % len(colors)])
    
    ax2.set_xlabel('Training Steps')
    ax2.set_ylabel('Average Reward')
    ax2.set_title('Reward Signal Over Time')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Reward variance (training stability)
    ax3 = fig.add_subplot(gs[0, 2])
    for i, result in enumerate(results):
        steps = [m.step for m in result.metrics]
        variances = [m.reward_variance for m in result.metrics]
        ax3.plot(steps, variances, marker='^', label=result.name, 
                linewidth=2, color=colors[i % len(colors)])
    
    ax3.set_xlabel('Training Steps')
    ax3.set_ylabel('Reward Variance')
    ax3.set_title('Training Stability (Lower = More Stable)')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 4. Perplexity over time
    ax4 = fig.add_subplot(gs[1, 0])
    for i, result in enumerate(results):
        steps = [m.step for m in result.metrics if m.perplexity > 0]
        perplexities = [m.perplexity for m in result.metrics if m.perplexity > 0]
        if steps:
            ax4.plot(steps, perplexities, marker='d', label=result.name, 
                    linewidth=2, color=colors[i % len(colors)])
    
    ax4.set_xlabel('Training Steps')
    ax4.set_ylabel('Perplexity')
    ax4.set_title('Model Perplexity Over Time')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    ax4.set_yscale('log')
    
    # 5. Reward distribution comparison
    ax5 = fig.add_subplot(gs[1, 1])
    for i, result in enumerate(results):
        ax5.hist(result.reward_distribution, alpha=0.6, label=result.name, 
                bins=30, color=colors[i % len(colors)], density=True)
    ax5.set_xlabel('Reward Value')
    ax5.set_ylabel('Density')
    ax5.set_title('Reward Distribution Comparison')
    ax5.legend()
    
    # 6. Learning efficiency (accuracy improvement per step)
    ax6 = fig.add_subplot(gs[1, 2])
    for i, result in enumerate(results):
        if len(result.metrics) > 1:
            steps = []
            efficiency = []
            for j in range(1, len(result.metrics)):
                step_diff = result.metrics[j].step - result.metrics[j-1].step
                acc_diff = result.metrics[j].accuracy - result.metrics[j-1].accuracy
                if step_diff > 0:
                    steps.append(result.metrics[j].step)
                    efficiency.append(acc_diff / step_diff)
            
            if steps:
                ax6.plot(steps, efficiency, marker='*', label=result.name, 
                        linewidth=2, color=colors[i % len(colors)])
    
    ax6.set_xlabel('Training Steps')
    ax6.set_ylabel('Learning Efficiency (Œî Accuracy / Œî Steps)')
    ax6.set_title('Learning Efficiency Over Time')
    ax6.legend()
    ax6.grid(True, alpha=0.3)
    
    # 7. Reward components breakdown (for sophisticated rewards)
    ax7 = fig.add_subplot(gs[2, 0])
    for i, result in enumerate(results):
        if result.metrics and result.metrics[0].reward_components:
            components = result.metrics[0].reward_components.keys()
            comp_means = []
            for comp in components:
                comp_values = []
                for metric in result.metrics:
                    if metric.reward_components and comp in metric.reward_components:
                        comp_values.append(metric.reward_components[comp])
                comp_means.append(np.mean(comp_values) if comp_values else 0)
            
            x_pos = np.arange(len(components)) + i * 0.2
            ax7.bar(x_pos, comp_means, width=0.2, label=result.name, 
                   color=colors[i % len(colors)], alpha=0.8)
    
    ax7.set_xlabel('Reward Components')
    ax7.set_ylabel('Average Component Value')
    ax7.set_title('Reward Components Breakdown')
    ax7.set_xticks(np.arange(len(components)) + 0.1)
    ax7.set_xticklabels(components, rotation=45)
    ax7.legend()
    
    # 8. Convergence comparison
    ax8 = fig.add_subplot(gs[2, 1])
    methods = [result.name for result in results]
    convergence_steps = [result.convergence_step for result in results]
    final_accuracies = [result.final_accuracy for result in results]
    
    scatter = ax8.scatter(convergence_steps, final_accuracies, 
                         c=range(len(results)), s=100, alpha=0.7, cmap='viridis')
    
    for i, method in enumerate(methods):
        ax8.annotate(method, (convergence_steps[i], final_accuracies[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax8.set_xlabel('Convergence Step')
    ax8.set_ylabel('Final Accuracy')
    ax8.set_title('Convergence Speed vs Final Performance')
    ax8.grid(True, alpha=0.3)
    
    # 9. Training stability vs performance
    ax9 = fig.add_subplot(gs[2, 2])
    stabilities = [result.training_stability for result in results]
    
    scatter = ax9.scatter(stabilities, final_accuracies, 
                         c=range(len(results)), s=100, alpha=0.7, cmap='plasma')
    
    for i, method in enumerate(methods):
        ax9.annotate(method, (stabilities[i], final_accuracies[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax9.set_xlabel('Training Stability')
    ax9.set_ylabel('Final Accuracy')
    ax9.set_title('Stability vs Performance Trade-off')
    ax9.grid(True, alpha=0.3)
    
    plt.suptitle('Comprehensive Training Dynamics Analysis', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('training_dynamics_comprehensive.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_reward_component_evolution(results: List[ExperimentResults]):
    """Plot evolution of individual reward components over time."""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Reward Component Evolution Over Time', fontsize=16)
    
    component_names = ['correctness', 'perplexity', 'reasoning_quality', 'consistency']
    
    for idx, comp_name in enumerate(component_names):
        ax = axes[idx // 2, idx % 2]
        
        for i, result in enumerate(results):
            if result.reward_type in ['continuous', 'hybrid']:
                steps = []
                comp_values = []
                
                for metric in result.metrics:
                    if metric.reward_components and comp_name in metric.reward_components:
                        steps.append(metric.step)
                        comp_values.append(metric.reward_components[comp_name])
                
                if steps:
                    ax.plot(steps, comp_values, marker='o', label=result.name, linewidth=2)
        
        ax.set_xlabel('Training Steps')
        ax.set_ylabel(f'{comp_name.title()} Score')
        ax.set_title(f'{comp_name.title()} Component Evolution')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('reward_components_evolution.png', dpi=300, bbox_inches='tight')
    plt.show()

def create_performance_heatmap(results: List[ExperimentResults]):
    """Create a heatmap comparing different performance metrics."""
    metrics_data = []
    method_names = []
    
    for result in results:
        method_names.append(result.name)
        # Normalize metrics to 0-1 scale for comparison
        metrics_data.append([
            result.final_accuracy,
            result.training_stability,
            1.0 / (1.0 + result.convergence_step / 10),  # Normalized convergence speed
            np.mean([m.avg_reward for m in result.metrics]) if result.metrics else 0,
            1.0 / (1.0 + np.mean([m.reward_variance for m in result.metrics])) if result.metrics else 0
        ])
    
    metrics_names = ['Final Accuracy', 'Training Stability', 'Convergence Speed', 
                    'Average Reward', 'Reward Consistency']
    
    # Create heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(metrics_data, 
                xticklabels=metrics_names,
                yticklabels=method_names,
                annot=True, 
                cmap='RdYlGn',
                fmt='.3f',
                cbar_kws={'label': 'Performance Score'})
    
    plt.title('Performance Metrics Comparison Heatmap')
    plt.tight_layout()
    plt.savefig('performance_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()

# ==================== ANALYSIS AND REPORTING ====================

def generate_comprehensive_analysis(results: List[ExperimentResults]) -> str:
    """Generate comprehensive analysis report."""
    report = []
    report.append("# Comprehensive RL Reward Structure Analysis")
    report.append("=" * 60)
    
    # Summary statistics
    report.append("\n## Summary Statistics")
    report.append("| Method | Final Accuracy | Convergence Step | Training Stability | Avg Reward |")
    report.append("|--------|----------------|------------------|-------------------|------------|")
    
    for result in results:
        avg_reward = np.mean([m.avg_reward for m in result.metrics]) if result.metrics else 0
        report.append(f"| {result.name} | {result.final_accuracy:.3f} | {result.convergence_step} | {result.training_stability:.3f} | {avg_reward:.3f} |")
    
    # Detailed analysis by reward type
    hard_results = [r for r in results if r.reward_type == 'hard']
    cont_results = [r for r in results if r.reward_type == 'continuous']
    hybrid_results = [r for r in results if r.reward_type == 'hybrid']
    
    report.append("\n## Analysis by Reward Type")
    
    if hard_results:
        report.append("\n### Hard Rewards Analysis")
        best_hard = max(hard_results, key=lambda x: x.final_accuracy)
        report.append(f"- Best performing hard reward method: {best_hard.name}")
        report.append(f"- Achieved {best_hard.final_accuracy:.3f} accuracy")
        report.append(f"- Converged at step {best_hard.convergence_step}")
        report.append("- Characteristics: Clear binary signal, fast convergence, potential training instability")
    
    if cont_results:
        report.append("\n### Continuous Rewards Analysis")
        best_cont = max(cont_results, key=lambda x: x.final_accuracy)
        report.append(f"- Best performing continuous reward method: {best_cont.name}")
        report.append(f"- Achieved {best_cont.final_accuracy:.3f} accuracy")
        report.append(f"- Converged at step {best_cont.convergence_step}")
        report.append("- Characteristics: Smooth learning signal, better stability, gradual improvement")
    
    if hybrid_results:
        report.append("\n### Hybrid Rewards Analysis")
        best_hybrid = max(hybrid_results, key=lambda x: x.final_accuracy)
        report.append(f"- Best performing hybrid reward method: {best_hybrid.name}")
        report.append(f"- Achieved {best_hybrid.final_accuracy:.3f} accuracy")
        report.append(f"- Converged at step {best_hybrid.convergence_step}")
        report.append("- Characteristics: Combines benefits of both approaches, adaptive learning")
    
    # Performance comparison
    report.append("\n## Key Findings")
    
    best_overall = max(results, key=lambda x: x.final_accuracy)
    most_stable = max(results, key=lambda x: x.training_stability)
    fastest_convergence = min(results, key=lambda x: x.convergence_step)
    
    report.append(f"1. **Highest Accuracy**: {best_overall.name} ({best_overall.final_accuracy:.3f})")
    report.append(f"2. **Most Stable Training**: {most_stable.name} (stability: {most_stable.training_stability:.3f})")
    report.append(f"3. **Fastest Convergence**: {fastest_convergence.name} (step {fastest_convergence.convergence_step})")
    
    # Statistical significance
    if len(results) >= 2:
        report.append("\n## Statistical Analysis")
        report.append("Performance differences analysis:")
        
        # Compare reward distributions
        from scipy import stats
        for i in range(len(results)):
            for j in range(i+1, len(results)):
                result1, result2 = results[i], results[j]
                
                if len(result1.reward_distribution) > 1 and len(result2.reward_distribution) > 1:
                    # T-test for reward distributions
                    t_stat, p_value = stats.ttest_ind(result1.reward_distribution, result2.reward_distribution)
                    
                    # Effect size (Cohen's d)
                    pooled_std = np.sqrt((np.var(result1.reward_distribution) + np.var(result2.reward_distribution)) / 2)
                    cohens_d = (np.mean(result1.reward_distribution) - np.mean(result2.reward_distribution)) / pooled_std
                    
                    significance = "**Significant**" if p_value < 0.05 else "Not significant"
                    effect_size = "Large" if abs(cohens_d) > 0.8 else "Medium" if abs(cohens_d) > 0.5 else "Small"
                    
                    report.append(f"- {result1.name} vs {result2.name}:")
                    report.append(f"  - P-value: {p_value:.4f} ({significance})")
                    report.append(f"  - Effect size: {cohens_d:.3f} ({effect_size})")
    
    # Recommendations
    report.append("\n## Recommendations")
    
    # Rank results by weighted score
    def calculate_weighted_score(result):
        return (0.4 * result.final_accuracy + 
                0.3 * result.training_stability + 
                0.3 * (1.0 / (1.0 + result.convergence_step / 10)))
    
    ranked_results = sorted(results, key=calculate_weighted_score, reverse=True)
    
    report.append(f"\n**Primary Recommendation: {ranked_results[0].name}**")
    report.append(f"- Overall score: {calculate_weighted_score(ranked_results[0]):.3f}")
    report.append("- Reasoning:")
    
    top_result = ranked_results[0]
    if top_result.reward_type == 'hard':
        report.append("  * Simple and effective binary rewards")
        report.append("  * Clear optimization target")
        report.append("  * Fast convergence when working well")
    elif top_result.reward_type == 'continuous':
        report.append("  * Smooth learning dynamics")
        report.append("  * Better handling of near-correct answers")
        report.append("  * More stable training process")
    else:  # hybrid
        report.append("  * Best of both approaches")
        report.append("  * Adaptive reward schedule")
        report.append("  * Balanced performance across metrics")
    
    report.append("\n**Use Cases by Reward Type:**")
    report.append("- **Hard Rewards**: When you have clear correct/incorrect answers and want fast convergence")
    report.append("- **Continuous Rewards**: When partial credit is valuable and training stability is important")
    report.append("- **Hybrid Rewards**: When you want to balance exploration and exploitation over training")
    
    return "\n".join(report)

# ==================== MAIN EXPERIMENT FUNCTION ====================

def main():
    """Run complete comparison experiment with all reward types."""
    print("Starting Enhanced RL Reward Structure Comparison")
    print("=" * 70)
    
    # Load GSM8K datasets
    print("Loading GSM8K datasets...")
    train_dataset = get_gsm8k_questions("train", limit=300)  # Reasonable size for experimentation
    eval_dataset = get_gsm8k_questions("test", limit=50)
    
    print(f"Loaded {len(train_dataset)} training examples and {len(eval_dataset)} evaluation examples")
    
    # Define all experiment configurations
    experiments = [
        get_hard_reward_config(),
        get_continuous_reward_config(),
        get_hybrid_reward_config("continuous_to_hard"),
        get_hybrid_reward_config("hard_to_continuous"),
    ]
    
    print(f"\nRunning {len(experiments)} experiments:")
    for exp in experiments:
        print(f"- {exp['name']}: {exp['description']}")
    
    # Run experiments
    results = []
    for i, config in enumerate(experiments):
        print(f"\n{'-'*50}")
        print(f"Experiment {i+1}/{len(experiments)}")
        try:
            result = run_experiment(config, train_dataset, eval_dataset)
            results.append(result)
            
            # Save intermediate results
            with open(f'experiment_{config["name"]}_results.json', 'w') as f:
                json.dump({
                    'name': result.name,
                    'final_accuracy': result.final_accuracy,
                    'convergence_step': result.convergence_step,
                    'training_stability': result.training_stability,
                    'reward_type': result.reward_type,
                    'metrics': [
                        {
                            'step': m.step,
                            'accuracy': m.accuracy,
                            'avg_reward': m.avg_reward,
                            'reward_variance': m.reward_variance,
                            'perplexity': m.perplexity,
                            'reward_components': m.reward_components
                        } for m in result.metrics
                    ]
                }, f, indent=2)
            
            print(f"‚úì Completed {config['name']}")
            print(f"  Final accuracy: {result.final_accuracy:.3f}")
            print(f"  Convergence step: {result.convergence_step}")
            
        except Exception as e:
            print(f"‚úó Failed {config['name']}: {str(e)}")
            continue
    
    if not results:
        print("No experiments completed successfully!")
        return []
    
    print(f"\n{'='*70}")
    print("GENERATING ANALYSIS AND VISUALIZATIONS")
    print("="*70)
    
    # Generate all visualizations
    print("Creating training dynamics visualizations...")
    visualize_training_dynamics(results)
    
    print("Creating reward component evolution plots...")
    plot_reward_component_evolution(results)
    
    print("Creating performance heatmap...")
    create_performance_heatmap(results)
    
    # Generate comprehensive analysis
    print("Generating comprehensive analysis report...")
    analysis_report = generate_comprehensive_analysis(results)
    
    # Save full report
    with open('comprehensive_rl_analysis.md', 'w') as f:
        f.write(analysis_report)
    
    # Print summary results
    print(f"\n{'='*70}")
    print("EXPERIMENT RESULTS SUMMARY")
    print("="*70)
    print(analysis_report)
    
    # Additional insights
    print(f"\n{'='*70}")
    print("KEY EXPERIMENTAL INSIGHTS")
    print("="*70)
    
    best_accuracy = max(results, key=lambda x: x.final_accuracy)
    most_stable = max(results, key=lambda x: x.training_stability)
    fastest = min(results, key=lambda x: x.convergence_step)
    
    print(f"\nüèÜ PERFORMANCE WINNERS:")
    print(f"   Highest Accuracy: {best_accuracy.name} ({best_accuracy.final_accuracy:.3f})")
    print(f"   Most Stable: {most_stable.name} (stability: {most_stable.training_stability:.3f})")
    print(f"   Fastest Convergence: {fastest.name} ({fastest.convergence_step} steps)")
    
    print(f"\nüìä REWARD TYPE ANALYSIS:")
    hard_results = [r for r in results if r.reward_type == 'hard']
    cont_results = [r for r in results if r.reward_type == 'continuous']
    hybrid_results = [r for r in results if r.reward_type == 'hybrid']
    
    if hard_results:
        avg_hard_acc = np.mean([r.final_accuracy for r in hard_results])
        print(f"   Hard Rewards Average Accuracy: {avg_hard_acc:.3f}")
    
    if cont_results:
        avg_cont_acc = np.mean([r.final_accuracy for r in cont_results])
        print(f"   Continuous Rewards Average Accuracy: {avg_cont_acc:.3f}")
    
    if hybrid_results:
        avg_hybrid_acc = np.mean([r.final_accuracy for r in hybrid_results])
        print(f"   Hybrid Rewards Average Accuracy: {avg_hybrid_acc:.3f}")
    
    print(f"\nüîÑ TRAINING DYNAMICS OBSERVATIONS:")
    for result in results:
        if result.metrics:
            initial_acc = result.metrics[0].accuracy
            final_acc = result.metrics[-1].accuracy
            improvement = final_acc - initial_acc
            print(f"   {result.name}: {initial_acc:.3f} ‚Üí {final_acc:.3f} (+{improvement:.3f})")
    
    print(f"\nüìÅ GENERATED FILES:")
    print("   - comprehensive_rl_analysis.md (detailed analysis)")
    print("   - training_dynamics_comprehensive.png (main visualizations)")
    print("   - reward_components_evolution.png (component analysis)")
    print("   - performance_heatmap.png (performance comparison)")
    print("   - experiment_*_results.json (raw data)")
    
    return results

# ==================== CREATIVE DOMAIN EXTENSION ====================

def run_creative_domain_experiments():
    """Run experiments on creative domains (poetry, jokes) as follow-up."""
    print(f"\n{'='*70}")
    print("CREATIVE DOMAIN EXPERIMENTS (FOLLOW-UP)")
    print("="*70)
    
    # Creative reward functions for domains without verifiable answers
    def aesthetic_reward_func(prompts, completions, **kwargs) -> list[float]:
        """Reward function for creative domains based on aesthetic qualities."""
        domain = kwargs.get('domain', 'poetry')
        responses = [completion[0]['content'] for completion in completions]
        rewards = []
        
        for response in responses:
            score = 0.0
            
            # 1. Length appropriateness (0.25 points)
            word_count = len(response.split())
            if domain == "poetry":
                ideal_length = 50
            else:  # jokes
                ideal_length = 30
            
            length_score = max(0, 1 - abs(word_count - ideal_length) / ideal_length)
            score += length_score * 0.25
            
            # 2. Structural quality (0.25 points)
            if domain == "poetry":
                lines = response.split('\n')
                if len(lines) > 1:
                    score += 0.125  # Multi-line structure
                if any(len(line.split()) <= 10 for line in lines):
                    score += 0.125  # Some concise lines
            else:  # jokes
                if "?" in response or "!" in response:
                    score += 0.125
                if any(word in response.lower() for word in ["why", "what", "how"]):
                    score += 0.125
            
            # 3. Creativity indicators (0.25 points)
            unique_words = len(set(response.lower().split()))
            total_words = len(response.split())
            if total_words > 0:
                creativity_score = unique_words / total_words
                score += creativity_score * 0.25
            
            # 4. Fluency (perplexity-based if model available) (0.25 points)
            model = kwargs.get('model')
            tokenizer = kwargs.get('tokenizer')
            if model and tokenizer:
                perplexity = calculate_perplexity(model, tokenizer, response)
                fluency_score = 1.0 / (1.0 + np.log(max(perplexity, 1.1)))
                score += fluency_score * 0.25
            else:
                score += 0.125  # Default partial credit
            
            rewards.append(min(1.0, score))
        
        return rewards
    
    # Create datasets for creative domains
    poetry_prompts = [
        "Write a haiku about technology",
        "Compose a short poem about friendship",
        "Create verses about the changing seasons",
        "Write a poem about dreams and aspirations",
        "Craft a nature poem about forests"
    ] * 20
    
    jokes_prompts = [
        "Tell a programming joke",
        "Write a pun about food",
        "Create a clean joke about daily life",
        "Make a witty observation about weather",
        "Tell a joke about books"
    ] * 20
    
    creative_datasets = {
        'poetry': Dataset.from_dict({
            'prompt': [[
                {'role': 'system', 'content': 'You are a creative poet. Write beautiful, original poetry.'},
                {'role': 'user', 'content': prompt}
            ] for prompt in poetry_prompts],
            'domain': ['poetry'] * len(poetry_prompts),
            'answer': [None] * len(poetry_prompts)
        }),
        'jokes': Dataset.from_dict({
            'prompt': [[
                {'role': 'system', 'content': 'You are a comedian. Write funny, clean jokes.'},
                {'role': 'user', 'content': prompt}
            ] for prompt in jokes_prompts],
            'domain': ['jokes'] * len(jokes_prompts),
            'answer': [None] * len(jokes_prompts)
        })
    }
    
    print("Note: Creative domain experiments would require:")
    print("1. Specialized evaluation metrics (human judgment, style analysis)")
    print("2. Domain-specific reward functions (aesthetic, humor, creativity)")
    print("3. Alternative success metrics (engagement, coherence, originality)")
    print("4. Longer training periods for creative skill development")
    print("\nThis demonstrates the framework's extensibility to subjective domains.")
    
    return creative_datasets

if __name__ == "__main__":
    # Run the main GSM8K experiments
    results = main()
    
    # Run creative domain demonstration
    print(f"\n{'='*70}")
    print("DEMONSTRATING CREATIVE DOMAIN EXTENSIBILITY")
    print("="*70)
    
    creative_datasets = run_creative_domain_experiments()
    
    # Final summary and next steps
    print(f"\n{'='*70}")
    print("EXPERIMENT COMPLETION SUMMARY")
    print("="*70)
    
    if results:
        print(f"\n‚úÖ Successfully completed {len(results)} experiments:")
        for result in results:
            print(f"   ‚Ä¢ {result.name}: {result.final_accuracy:.3f} accuracy")
        
        # Determine overall winner
        best_overall = max(results, key=lambda x: x.final_accuracy)
        print(f"\nüèÜ OVERALL WINNER: {best_overall.name}")
        print(f"   Final Accuracy: {best_overall.final_accuracy:.3f}")
        print(f"   Convergence Step: {best_overall.convergence_step}")
        print(f"   Training Stability: {best_overall.training_stability:.3f}")
        
        # Key insights
        print(f"\nüí° KEY INSIGHTS:")
        
        # Compare reward types
        hard_results = [r for r in results if r.reward_type == 'hard']
        cont_results = [r for r in results if r.reward_type == 'continuous']
        hybrid_results = [r for r in results if r.reward_type == 'hybrid']
        
        if hard_results and cont_results:
            hard_avg = np.mean([r.final_accuracy for r in hard_results])
            cont_avg = np.mean([r.final_accuracy for r in cont_results])
            
            if cont_avg > hard_avg:
                improvement = ((cont_avg - hard_avg) / hard_avg) * 100
                print(f"   ‚Ä¢ Continuous rewards outperformed hard rewards by {improvement:.1f}%")
                print(f"   ‚Ä¢ Continuous rewards provide smoother learning dynamics")
                print(f"   ‚Ä¢ Partial credit mechanisms help with near-correct answers")
            else:
                improvement = ((hard_avg - cont_avg) / cont_avg) * 100
                print(f"   ‚Ä¢ Hard rewards outperformed continuous rewards by {improvement:.1f}%")
                print(f"   ‚Ä¢ Binary signals can provide clearer optimization targets")
                print(f"   ‚Ä¢ Simpler reward structures may be more effective")
        
        if hybrid_results:
            hybrid_avg = np.mean([r.final_accuracy for r in hybrid_results])
            print(f"   ‚Ä¢ Hybrid approaches achieved {hybrid_avg:.3f} average accuracy")
            print(f"   ‚Ä¢ Adaptive reward schedules can balance exploration/exploitation")
        
        # Training dynamics insights
        print(f"\nüìà TRAINING DYNAMICS INSIGHTS:")
        fastest_convergence = min(results, key=lambda x: x.convergence_step)
        most_stable = max(results, key=lambda x: x.training_stability)
        
        print(f"   ‚Ä¢ Fastest convergence: {fastest_convergence.name} ({fastest_convergence.convergence_step} steps)")
        print(f"   ‚Ä¢ Most stable training: {most_stable.name} (stability: {most_stable.training_stability:.3f})")
        
        # Practical recommendations
        print(f"\nüéØ PRACTICAL RECOMMENDATIONS:")
        
        if best_overall.reward_type == 'hard':
            print(f"   ‚Ä¢ Use hard rewards when:")
            print(f"     - You have clear correct/incorrect answers")
            print(f"     - Fast convergence is prioritized")
            print(f"     - Training data is abundant and high-quality")
        elif best_overall.reward_type == 'continuous':
            print(f"   ‚Ä¢ Use continuous rewards when:")
            print(f"     - Partial credit is meaningful")
            print(f"     - Training stability is important")
            print(f"     - Working with complex reasoning tasks")
        else:  # hybrid
            print(f"   ‚Ä¢ Use hybrid rewards when:")
            print(f"     - You want to balance multiple objectives")
            print(f"     - Adaptive learning is beneficial")
            print(f"     - Training resources allow for complexity")
    
    else:
        print("\n‚ùå No experiments completed successfully.")
        print("Check the error messages above for debugging information.")
    
    # Future work suggestions
    print(f"\nüîÆ FUTURE WORK SUGGESTIONS:")
    print("   1. Scale experiments to larger models (7B, 13B parameters)")
    print("   2. Test on diverse mathematical reasoning datasets (MATH, MathQA)")
    print("   3. Implement more sophisticated hybrid schedules")
    print("   4. Add human evaluation for creative domains")
    print("   5. Investigate reward normalization techniques")
    print("   6. Study few-shot vs fine-tuning performance differences")
    print("   7. Analyze computational efficiency of different reward types")
    
    # File summary
    print(f"\nüìã GENERATED OUTPUTS:")
    print("   Analysis Reports:")
    print("   ‚Ä¢ comprehensive_rl_analysis.md - Complete experimental analysis")
    print("   ‚Ä¢ experiment_*_results.json - Raw experimental data")
    print("")
    print("   Visualizations:")
    print("   ‚Ä¢ training_dynamics_comprehensive.png - Main training dynamics")
    print("   ‚Ä¢ reward_components_evolution.png - Component-wise analysis")
    print("   ‚Ä¢ performance_heatmap.png - Performance comparison matrix")
    print("")
    print("   Usage Instructions:")
    print("   ‚Ä¢ Review the markdown report for detailed findings")
    print("   ‚Ä¢ Use the JSON files for further analysis or replication")
    print("   ‚Ä¢ Examine the visualizations for training insights")
    
    # Reproducibility information
    print(f"\nüîÑ REPRODUCIBILITY:")
    print("   ‚Ä¢ Random seed: 3407 (set in model configuration)")
    print("   ‚Ä¢ Model: Qwen/Qwen2.5-3B-Instruct")
    print("   ‚Ä¢ Dataset: GSM8K (OpenAI)")
    print("   ‚Ä¢ Training steps: 15 per experiment")
    print("   ‚Ä¢ Evaluation frequency: Every 2 steps")
    
    print(f"\n{'='*70}")
    print("EXPERIMENT COMPLETED SUCCESSFULLY!")
    print("="*70)
    print("Thank you for using the Enhanced RL Reward Comparison Framework!")
    print("Check the generated files for detailed results and analysis.")
    print("For questions or issues, review the code comments and error messages.")
    print("="*70)

    # Optional: Create a quick summary JSON for programmatic access
    if results:
        summary = {
            "experiment_date": "2024",
            "total_experiments": len(results),
            "best_performer": {
                "name": best_overall.name,
                "accuracy": best_overall.final_accuracy,
                "reward_type": best_overall.reward_type
            },
            "reward_type_performance": {
                "hard": np.mean([r.final_accuracy for r in results if r.reward_type == 'hard']) if hard_results else None,
                "continuous": np.mean([r.final_accuracy for r in results if r.reward_type == 'continuous']) if cont_results else None,
                "hybrid": np.mean([r.final_accuracy for r in results if r.reward_type == 'hybrid']) if hybrid_results else None
            },
            "key_findings": [
                f"Best overall method: {best_overall.name}",
                f"Fastest convergence: {fastest_convergence.name}",
                f"Most stable training: {most_stable.name}"
            ]
        }
        
        with open('experiment_summary.json', 'w') as f:
            json.dump(summary, f, indent=2)
        
        print(f"\nüíæ Summary saved to: experiment_summary.json")
```
